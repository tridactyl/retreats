```python
categories = """
PayPal
Patreon
UK
USA
Germany
France
Netherlands
Denmark
Sweden
Norway
Italy
Switzerland
Accommodation
Donations
Funds
Transport
Taxi
Train
Bus
Aeroplane
Shopping
Alcohol
Groceries
Food
Restaurants
Eating-out
Voucher
Fees
Developers
""".split("\n")
edges = [
    "UK PayPal 35",
    "USA PayPal 246",
    "Germany PayPal 183",
    "Netherlands PayPal 60",
    "Denmark PayPal 50",
    "Sweden PayPal 40",
    "Norway PayPal 30",
    "Italy PayPal 24",
    "Switzerland PayPal 20",
    "PayPal Fees " + str(1028-180-811.54), # 180 was refunded
    "PayPal Donations 811.54",
    "Patreon Donations 3", # Not including countries as will de-anonymise people
    "Donations Funds 811.54", # Haven't withdrawn from Patreon
    "Funds Accommodation 1395.40",
    "Transport Train " + str(12*2 + 25*2),
    "Transport Taxi " + str(5*2),
    "Transport Bus " + str(4),
    "Transport Aeroplane " + str(300 + 212), # unsure how much other flight was
    "Funds Transport 600",
    "Funds Food " + str(143+130+15),
    "Food Shopping 143",
    "Shopping Groceries " + str(143-36),
    "Shopping Alcohol 36",
    "Food Eating-out " + str(130+15),
    "Voucher Aeroplane " + str(212-30),
    "Developers Funds " + str(2280-811.54),
    "Developers PayPal 70"
]



links = {
    targ: list(map(lambda x: x[targ], [
        {'source': categories.index(x[0]), 'target': categories.index(x[1]), 'value': x[2]} for x in map(lambda x: x.split(), edges)
    ])) for targ in ["source","target","value"]
}


data = dict(
    type='sankey',
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(
        color = "black",
        width = 0.5
      ),
      label = categories
#      color = ["blue", "blue", "blue", "blue", "blue", "blue"]
    ),
    link = links
)

layout =  dict(
    title = "Tridactyl 2019 retreat funding and expenditure (prices in sterling)",
    font = dict(
      size = 10
    )
)

fig = dict(data=[data], layout=layout)
iplot(fig, validate=False)
```
